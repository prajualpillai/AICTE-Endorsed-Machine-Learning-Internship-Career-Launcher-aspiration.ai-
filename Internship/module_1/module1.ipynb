## 1.4

To calculate the monthwise VWAP (Volume Weighted Average Price) of the stock.

```python
df_1['Year'] = pd.DatetimeIndex(df_1['Date']).year
df_1['Month'] = pd.DatetimeIndex(df_1['Date']).month
```

```python
df_3 = df_1.groupby('Year')
df_3.first() # print first entries of all the years formed
```

```python
for i in range(3):
    year = 2017 + i
    df_temp_year = df_3.get_group(year)
    df_temp_months = df_temp_year.groupby('Month')
    month_vals = list(set(df_temp_year['Month'].values))
    for j in month_vals:
        df_temp_month = df_temp_months.get_group(j)
        vwap = sum(df_temp_month['Close Price'] * df_temp_month['Total Traded Quantity'])/sum(df_temp_month['Total Traded Quantity'])
        print(f"VWAP for Year {year} and Month {j} : {vwap:.2f}")
```

## 1.5

#### To calculate average price and Profit/Loss of last N days/weeks/months/year of the stock.

```python
df_3 = df_1.set_index(['Date']) #setting indexing w.r.t datetime
```

```python
print("==To calculate average price and Profit/Loss of last N days/weeks/months/year of the stock==")
choice = str(input("Choose:\nDay - D\nWeeks - W\nMonths - M\nYear - Y\n"))
if choice.lower() == 'd':
    choice_word = 'Days'
    c = 'd'
elif choice.lower() == 'w':
    choice_word = 'Weeks'
    c = 'w'
elif choice.lower() == 'm':
    choice_word = 'Months'
    c ='d'
elif choice.lower() == 'y':
    choice_word = 'Years'
    c = 'd'
else:
    print("Wrong Choice")
if choice.lower() in ['d','w','m','y']:
    x = int(input(f"Enter Number of {choice_word}: "))
if choice.lower() == 'm':
    x = x*30
if choice.lower() == 'y':
    x = x*365
```

```python
start_date = df_1['Date'].max() - pd.to_timedelta(x, unit=c)
print(f"Start Date: {start_date.date()}")
print(f"End Date: {str(df_1['Date'].max().date())}")
```

```python
df_diff_n = df_3.loc[str(start_date.date()):str(df_1['Date'].max().date())]
```

```python
avg_price_n = df_diff_n["Average Price"].sum()/df_diff_n["Average Price"].count()
```

```python
d_1 = df_diff_n[:1]
d_n = df_diff_n[-1:]
profit_loss = 100*(d_n["Close Price"].values[0] - d_1["Close Price"].values[0])/d_1["Close Price"].values[0]
```

```python
print(f"Average Price for past {x} {choice_word}: {avg_price_n:.2f}")
print(f"Profit/Loss percentage for past {x} {choice_word}: {profit_loss:.2f}%")
```

## 1.6

To find percentage change btw 2 consecutive day's closing price.

```python
df_4 = df_1.copy()
```

```python
df_4["Day_Perc_Change"] = df_4["Close Price"].pct_change(periods = 1)
df_4.head()
```

NaN comes in first row because there is no last valid observation to caluclate the percentage change for the first row. Also if we change the period from 1 to 2, then two NaN will appear in the first two rows due to the same reason.

```python
df_4 = df_4.iloc[1:] #removing the first row
df_4.head()
```

## 1.7

```python
df_5 = df_4.copy()
```

```python
df_5["Trend"] = np.nan #create a new empty column in dataframe
```

```python
for index, row in df_5.iterrows():
    if row["Day_Perc_Change"] > -0.5 and row["Day_Perc_Change"] < 0.5:
        df_5.loc[index, "Trend"] = 'Slight/NoChange'
    if row["Day_Perc_Change"] > 0.5 and row["Day_Perc_Change"] < 1:
        df_5.loc[index, "Trend"] = 'Slight Positive'
    if row["Day_Perc_Change"] < -0.5 and row["Day_Perc_Change"] > -1:
        df_5.loc[index, "Trend"] = 'Slight Negative'
    if row["Day_Perc_Change"] > 1 and row["Day_Perc_Change"] < 3:
        df_5.loc[index, "Trend"] = 'Positive'
    if row["Day_Perc_Change"] > -3 and row["Day_Perc_Change"] < -1:
        df_5.loc[index, "Trend"] = 'Negative'
    if row["Day_Perc_Change"] > 3 and row["Day_Perc_Change"] < 7:
        df_5.loc[index, "Trend"] = 'Among top gainers'
    if row["Day_Perc_Change"] < -3 and row["Day_Perc_Change"] > -7:
        df_5.loc[index, "Trend"] = 'Among top losers'
    if row["Day_Perc_Change"] > 7:
        df_5.loc[index, "Trend"] = 'Bull run'
    if row["Day_Perc_Change"] < -7:
        df_5.loc[index, "Trend"] = 'Bear drop'
```

To find average and median values of colum 'Total Traded Quantity' for each type of 'Trend'

```python
df_6 = df_5.copy()
```

```python
df_trend_groups = df_6.groupby('Trend')
```

```python
df_trend_groups.first()
```

```python
df_slight_negative = df_trend_groups.get_group('Slight Negative')
```

```python
for group in list(set(df_6['Trend'].values)):
    df_group = df_trend_groups.get_group(group)
    avg = df_group['Total Traded Quantity'].mean()
    median = df_group['Total Traded Quantity'].median()
    print(f"For Trend '{group}'")
    print(f"Average Total Traded Quantity: {avg:.2f}")
    print(f"Median Total Traded Quantity: {median:.2f}")
    print("-"*20)
```
